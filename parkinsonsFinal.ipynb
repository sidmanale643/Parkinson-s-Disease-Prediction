```python
X = df.drop(['status'] , axis = 1)  
y = df['status']
```

```python
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
```

```python
scaler = StandardScaler()
xgb_classifier = xgb.XGBClassifier(learning_rate = 0.001)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2)
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
np.mean(scaled_X_train)
```

```python
np.std(scaled_X_train)
```

```python
X_test.shape
```

```python
plt.figure(figsize = (25,10))
sns.boxplot(scaled_X_train)
plt.tight_layout()
```

```python
from sklearn.metrics import accuracy_score
```

```python
xgb_classifier.fit(scaled_X_train , y_train)
train_pred = xgb_classifier.predict(scaled_X_train)
acc = accuracy_score(y_train , train_pred)
```

```python
print(f"The training accuracy of the XGBoost Classifier model is {acc * 100}%")
```

```python
test_pred = xgb_classifier.predict(scaled_X_test)
test_acc = accuracy_score(y_test , test_pred)
```
